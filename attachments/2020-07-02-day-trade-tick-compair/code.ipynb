```python
import pandas as pd
import numpy as np
```

```python
price=np.arange(0,1100)
```

```python
df = pd.DataFrame(columns=['price','h_fee','back_fee','tick','start_make'])
```

```python
df
```

```python
df['price']=price
```

```python
df['tick']=df['price'].apply(lambda a:0.01 if a <10 
                             else (
                             0.05 if a>=10 and a<50
                             else(
                             0.1 if a>=50 and a<100 
                             else(
                             0.5 if a>=100 and a<500
                             else(
                             1 if a>=500 and a<1000
                             else (
                             5 if a>=1000 else 0)))))
                            )
```

```python
percent = 0.65
discount = (0.1425*(1-percent))*2
fee = 0.1425 + 0.1425 + 0.15

df['h_fee']=(df['price']*1000*fee)/100
df['back_fee'] = (df['price']*1000*discount)/100
df['start_make']=df['h_fee']/(df['tick']*1000)
```
